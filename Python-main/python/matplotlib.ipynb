```python
import matplotlib.pyplot as plt
plt.rc('font',family='Gulim')
x=[1,2,3,4]
y=[10,4,15,9]
plt.plot(x,y,label='국어')
plt.title('그래프 예제')
plt.xlabel('번호')
plt.ylabel('점수')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips_data=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',hue='day',data=tips_data)
plt.title('Total bill & tips')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips_data=sns.load_dataset('tips')
sns.barplot(x='day',y='tip',data=tips_data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips_data=sns.load_dataset('tips')
sns.histplot(x='tip',data=tips_data,kde=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_excel('data3.xlsx')
plt.rc('font',family='Gulim')

x=['김하나','이정희','최선영','박영준','송정민']
plt.plot(df['name'],df['kor'],'k--',marker='o',label='국어점수')
plt.plot(df['name'],df['math'],marker='v',label='수학점수')
plt.title('성적 그래프')
plt.xlabel('이름')
plt.ylabel('점수')
plt.legend()
plt.show()
```
