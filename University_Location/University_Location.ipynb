```python
import googlemaps
import pandas as pd
import folium
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

mykey = "AIzaSyB0HPuOS6EDcIEjxLIk7u08dtrkI98cGfQ"
maps = googlemaps.Client(key=mykey)

mymap=folium.Map(location=[36, 127],zoom_start=7)

df=pd.read_excel('고등교육기관 하반기 주소록(2022).xlsx')

def get_addrs(x):
    x1 = x.split(' ')
    return " ".join(x1[0:4])

df['도로명주소'] = df['주소'].apply(get_addrs)

def geocoding_lat(addr):
    try:
        geo_location = maps.geocode(addr)[0].get('geometry')
        lat = geo_location['location']['lat']
        return lat
    except:
        return 0
    
def geocoding_lon(addr):
    try:
        geo_location = maps.geocode(addr)[0].get('geometry')
        lng =  geo_location['location']['lng']
        return lng
    except:
        return 0    

df['위도'] = df['도로명주소'].apply(geocoding_lat)
df['경도'] = df['도로명주소'].apply(geocoding_lon)

for i in range(len(df)):
    folium.Marker([float(df['위도'][i:i+1]),float(df['경도'][i:i+1])],popup=df['학교명'][i:i+1]).add_to(mymap)
    
for i in range(len(df)):
    folium.Marker([float(df.iloc[i,5]),float(df.iloc[i,6])],popup=df.iloc[i,2]).add_to(mymap)
mymap
```
